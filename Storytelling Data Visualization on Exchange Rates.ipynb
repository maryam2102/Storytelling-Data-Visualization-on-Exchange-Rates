# Storytelling Data Visualization on Exchange Rates

The dataset contains the exchange rate data for various countries spanning the years 1999 to 2023.. This project explores how the Euro's value relative to the US Dollar evolved over the years. We'll examine the fluctuations in the exchange rate and analyze any discernible patterns that emerge over different time periods.

```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
#The dataset contains the data about the exchange rates of Euros to other Currencies from 1999 to 2023
exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")
exchange_rates.head()
```

```python
exchange_rates.shape
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns = {'[US dollar ]' : 'US_dollar' , 'Period\\Unit:' : 'Time'} , inplace = True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time' , inplace = True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time' , 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() #62 "-" symbols
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

 ## Rolling Mean

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar['Time'] , euro_to_dollar['US_dollar'])
plt.text(0.5 , -0.2 , """If we look at the line's shape, we see many small wiggles rather than a smooth line. 
    The wiggles basically are showing the daily variation in the exchange rates. 
    The rate goes up and down again and again, day by day. 
    The rate shows the clear upward or downward trends only in the longer run (months or years). """, 
    horizontalalignment = 'center' , verticalalignment = 'center', fontsize = 12 , 
    transform = plt.gca().transAxes , color = 'black')
plt.show()
```

#### Depending on the requirements, we may not want to show the daily variation in our graph. If we want to show only the long-term trends, we can use the rolling mean (also known as moving average)

#### A moving average shows the average value of something over time. It helps to get rid of bumps or sudden changes in data, making it easier to see the overall trend.

```python
values = pd.DataFrame()
values['daily_values'] = pd.Series(range(1,20,2))
values
```

```python
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
values
```

```python
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()
values
```

```python
plt.figure(figsize=(9,6))


plt.subplot(2,2,1)
plt.plot(euro_to_dollar['Time'] , euro_to_dollar['US_dollar'])
plt.title('Original Values' , weight = 'bold')
plt.ylim(0.7,1.8,0.2)

for i , rolling_mean in zip([2,3,4] , [7,30,365]): 
    plt.subplot(2,2,i)
    plt.plot(euro_to_dollar['Time'] , 
            euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window :' + str(rolling_mean) , weight = 'bold')
    plt.ylim(0.7,1.8,0.2)

plt.tight_layout()
plt.show()
```

In the plotted data, the graph representing a Rolling Mean of 365 days offers a clearer depiction of the overall trend in the exchange rate over time. This extended rolling mean effectively smooths out fluctuations, allowing for more insightful observations regarding the rate's sustained rise or fall.

Comparatively, the graphs depicting monthly or weekly rolling means exhibit more fluctuations or "wiggles," indicating rapid changes in the exchange rate within shorter time frames. These fluctuations can obscure the underlying trend, making it challenging to discern accurate insights from them.

#### In the graph window of 365 days, it can be depicted clearly that the euro-dollar exchange rates were going down between 2001 to 2003, 

#### whereas between 2008 to 2010, the euro-dollar exchange rates were at the highest peak. 

# StoryTelling and Visualization

## Financial Crisis 2007-2008: 

#### Now, the graph is plotted to understand how did the euro-dollar exchange rate change during the financial crisis of 2007-2008 

#### and we'll be using the data of 2006-2009 as a baseline

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) 
                                        & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) 
                                        & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

#Plot
fig,ax = plt.subplots(figsize = (8,3))
ax.plot(financial_crisis['Time'] , financial_crisis['rolling_mean'] , 
        linewidth = 1.5 , color = '#898f82')

#Highlighting 2007-2008 Period
ax.plot(financial_crisis_7_8['Time'] , financial_crisis_7_8['rolling_mean'],
       linewidth = 3 , color = '#ad0715')

#Clearing x labels
ax.set_xticklabels([])

#Setting x labels
x = 0.02
for year in ['2006','2007','2008','2009','2010']:
    ax.text(x , -0.07 , year , alpha = 0.5 , fontsize = 11 , transform = plt.gca().transAxes)
    x += 0.228

#Clearing y labels
ax.set_yticklabels([])

#Setting y labels
y = 0.07
for rate in ['1.2','1.3','1.4','1.5']:
    ax.text(-0.04 , y , rate , alpha = 0.5 , fontsize = 11 , transform = plt.gca().transAxes)
    y += 0.23

# Adding a Title and a Subtitle
ax.text(-0.05 , 1.2 , "Euro-USD Exchange Rate Peaked at 1.59 During 2007-2008's Financial Crisis",
       weight = 'bold' , transform = plt.gca().transAxes)

ax.text(-0.05 , 1.1 , "Euro-USD Exchange Rate between 2006 and 2010",
        size = 11 , transform = plt.gca().transAxes)

#Adding a Signature
ax.text(-0.05 , -0.25 , '@TECHMA ZONE' + " "*80 + 'Source: European Central Bank',
       color = '#eae9f2' , backgroundcolor = '#858487' , size = 12 , transform = plt.gca().transAxes)

#Highlighting the Peak Area
ax.axvspan(xmin = pd.to_datetime('2008-04-1') , xmax = pd.to_datetime('2008-09-1') , ymin = 0.09 , 
          alpha = 0.3 , color = 'grey')
```

## Coronavirus Pandemic: 

#### Now, the graph is plotted to understand how did the euro-dollar exchange rate change during the covid-19 .

#### We'll be showing the data of 2020 and using the data of 2016-2022 as a baseline

```python
corona_crisis_20 = euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2020-01-01') & 
                                     (euro_to_dollar['Time'] <= '2020-12-31')]

corona_crisis = euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2016-01-1') & 
                                     (euro_to_dollar['Time'] <= '2022-12-31')]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

#Plot
fig,ax = plt.subplots(figsize = (10,3))
ax.plot(corona_crisis['Time'] , corona_crisis['rolling_mean'] , 
        linewidth = 2 , color = '#898f82' , alpha = 0.5)

#Highlighting 2007-2008 Period
ax.plot(corona_crisis_20['Time'] , corona_crisis_20['rolling_mean'],
       linewidth = 3 , color = '#ad0715')


#Clearing x labels
ax.set_xticklabels([])

#Setting x labels
x = 0.02
for year in ['2016','2017','2018','2019','2020','2021','2022','2023']:
    ax.text(x , -0.08 , year , alpha = 0.5 , fontsize = 11 , transform = plt.gca().transAxes)
    x += 0.13

#Clearing y labels
ax.set_yticklabels([])

#Setting y labels
y = 0.02
for rate in ['1.05','1.10','1.15','1.20']:
    ax.text(-0.04 , y , rate , alpha = 0.5 , fontsize = 11 , transform = plt.gca().transAxes)
    y += 0.248

# Adding a Title and a Subtitle
ax.text(-0.05 , 1.2 , "Euro-USD Exchange Rate Peaked at 1.22 During Covid19 Crisis",
       weight = 'bold' , transform = plt.gca().transAxes)

ax.text(-0.05 , 1.1 , "Euro-USD Exchange Rate between 2019 and 2023",
        size = 11 , transform = plt.gca().transAxes)

#Adding a Signature
ax.text(-0.05 , -0.25 , '@TECHMA ZONE' + " "*96 + 'Source: European Central Bank',
       color = '#eae9f2' , backgroundcolor = '#858487' , size = 12 , transform = plt.gca().transAxes)


ax.axvspan(xmin=pd.to_datetime("2020-01-1"), xmax=pd.to_datetime("2020-12-31"),ymin=0.07,ymax=0.9, color='#ebe534',alpha=0.5)

plt.show()
```

## The Three US Presidencies

Now, the graph will be plotted to show comparatively how did the Euro-Dollar exchange rate change during different presidencies 

of (George W.Bush (2001-2009) , Barack Obama (2009-2017) and Donald Trump (2017-2021)). 

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & 
                                         (euro_to_dollar['Time'].dt.year <= 2021) ]

bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]

obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2009) & 
                               (bush_obama_trump['Time'].dt.year < 2017)]

trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & 
                               (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
style.use('fivethirtyeight')
#Adding Subplots
plt.figure(figsize = (14,9))

#pattern 1
ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

#pattern 2
ax4 = plt.subplot(3,1,2)

axes = [ax1 , ax2 , ax3 , ax4]

#Setting labels for all the subplots
for ax in axes:
    ax.set_ylim(0.8 , 1.7)
    ax.set_yticks([1.0 , 1.2 , 1.4 , 1.6])
    ax.set_yticklabels(['1.0' , '1.2' , '1.4' , '1.6$'] , alpha = 0.5)

#Ax1 : Bush
ax1.plot(bush['Time'] , bush['rolling_mean'] , color = '#36c9a5' )
ax1.set_xticklabels(['' , '2001' , '' , '2003' , '' ,'2005' , '' ,'2007' , '' ,'2009'] , alpha = 0.5 , size = 12)
ax1.text(0.11 , 2.45 , 'BUSH' , fontsize = 20 , weight = 'bold' , color = '#FF7F50' , transform = plt.gca().transAxes)
ax1.text(0.093 , 2.34 , '(2001-2009)' , weight = 'bold' , alpha = 0.3 , transform = plt.gca().transAxes)

#Ax2 : Obama
ax2.plot(obama['Time'] , obama['rolling_mean'] , color = '#ab37bf')
ax2.set_xticklabels(['' , '2001' , '' , '2003' , '' ,'2005' , '' ,'2007' , '' ,'2009'] , alpha = 0.5 , size = 12)
ax2.text(0.45 , 2.45 , 'OBAMA' , fontsize = 20 , weight = 'bold' , color = '#008080' , transform = plt.gca().transAxes)
ax2.text(0.44 , 2.34 , '(2009-2017)' , weight = 'bold' , alpha = 0.3 , transform = plt.gca().transAxes)

#Ax3 : Trump
ax3.plot(trump['Time'] , trump['rolling_mean'] , color = '#f02e85')
ax3.set_xticklabels(['' , '2001' , '' , '2003' , '' ,'2005' , '' ,'2007' , '' ,'2009'] , alpha = 0.5 , size = 12)
ax3.text(0.82 , 2.45 , 'TRUMP' , fontsize = 20 , weight = 'bold' , color = '#db604b' , transform = plt.gca().transAxes)
ax3.text(0.808 , 2.34 , '(2017-2021)' , weight = 'bold' , alpha = 0.3 , transform = plt.gca().transAxes)

#Ax4 : Bush , Obama , Trump
ax4.plot(bush['Time'] , bush['rolling_mean'] , color = '#36c9a5' )
ax4.plot(obama['Time'] , obama['rolling_mean'] , color = '#ab37bf' )
ax4.plot(trump['Time'] , trump['rolling_mean'] , color = '#f02e85' )

ax4.set_xticks([])

#Adding a title
ax1.text(-0.05 , 2.7 , """EURO-USD Exchange Rate Under George W. Bush (2001-2009), Barack Obama (2009-2017) and Donald Trump (2017-2021)""" ,
          fontsize = 14, weight = 'bold' , color = '#5d1f8c' , transform = plt.gca().transAxes)

ax1.text(0.05 , -0.1 , """EURO-USD Exchange Rate was average 1.22 under the last three US Presidents""" ,
          fontsize = 16, weight = 'bold' , color = '#eb5754' , transform = plt.gca().transAxes)

ax.text(-0.05 , -0.25 , '@TECHMA ZONE' + " "*190 + 'Source: European Central Bank',
       color = '#eae9f2' , backgroundcolor = '#858487' , size = 12 , transform = plt.gca().transAxes)

plt.show()







```
